```python
#Import basic librarys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Select input and output from dataset
X=df.drop("Class",axis=1)#input
Y=df["Class"]#output
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Use randomoversampling
from imblearn.over_sampling import RandomOverSampler
```

```python
#RandomOverSampler
#create object of RandomOverSampler() class
ros=RandomOverSampler(random_state=1)
```

```python
X_train_ros,Y_train_ros=ros.fit_resample(X_train,Y_train)
#fit_resample() inbuild method of RandomOverSampler class
```

```python
pd.Series(Y_train).value_counts()
```

```python
pd.Series(Y_train_ros).value_counts()
```

```python
#Apply randomoversampling on 30% testing data
X_test_ros,Y_test_ros=ros.fit_resample(X_test,Y_test)
```

```python
pd.Series(Y_test).value_counts()
```

```python
pd.Series(Y_test_ros).value_counts()
```

```python
def create_model(model):
  model.fit(X_train_ros,Y_train_ros)
  Y_pred=model.predict(X_test_ros)
  print(classification_report(Y_test_ros,Y_pred))
  print(confusion_matrix(Y_test_ros,Y_pred))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```
